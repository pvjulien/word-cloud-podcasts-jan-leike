```python
# Install the required library if not already installed
# !pip install wordcloud

# Import necessary modules
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Read text from the "transcripts.txt" file
with open("transcripts.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Specify the maximum number of words
max_words = 75  # Change this value to the desired maximum number of words
```

```python
# Read custom stopwords from an Excel file
custom_stopwords_file = "custom_stopwords.xlsx"
custom_stopwords_df = pd.read_excel(custom_stopwords_file)
custom_stopwords = set(custom_stopwords_df["Stopword"])

# Combine custom stopwords with default stopwords
combined_stopwords = custom_stopwords.union(STOPWORDS)
```

```python
# Create a WordCloud object with stopwords and horizontal layout
wordcloud = WordCloud(width=1200, height=800,
                      background_color='white',
                      stopwords=combined_stopwords,  # Use combined stopwords set
                      min_font_size=20,
                      max_font_size=180,
                      colormap='tab10',
                      max_words=max_words,
                      collocations=True,
                      collocation_threshold=3,
                      prefer_horizontal=1).generate(text)  # Set horizontal_layout to True
```
